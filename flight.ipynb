```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_excel(r"data/Data_Train.xlsx")
```

```python
train_data.head()
```

```python
train_data.shape
```

Pre-Processing

```python
train_data.dropna(inplace = True)
```

```python
#Extract the data of date and month of the journey
train_data["Journey_day"] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
train_data["Journey_month"] = pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
#Drop the Date_of_Journey parameter
train_data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
#Extract the hours and minute of departure time
train_data["Dep_hour"] = pd.to_datetime(train_data.Dep_Time).dt.hour
```

```python
train_data["Dep_min"] = pd.to_datetime(train_data.Dep_Time).dt.minute
```

```python
train_data.drop(["Dep_Time"], axis = 1 , inplace = True)
```

```python
train_data.head()
```

```python
#Now extract hour and minute data from Arrival_Time column
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute
```

```python
train_data.drop(["Arrival_Time"], axis = 1 , inplace = True)
```

```python
train_data.head()
```

```python
#Assigning and converting Duration column into list
duration = list(train_data.Duration)

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i]  = duration[i].strip() + " 0m"
        else:
            duration[i] = "0h " + duration[i]

duration_hours = []
duration_min = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    duration_min.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_min
```

```python
train_data.drop(["Duration"], axis = 1, inplace = True)
```

```python
train_data.drop(["Route","Additional_Info"], axis = 1, inplace = True )
```

```python
train_data.head()
```

```python
train_data.replace({"non-stop":0 , "1 stop": 1, "2 stops" : 2 , "3 stops" : 3, "4 stops":4}, inplace = True)
```

```python
train_data.head()
```

```python
# Perform OneHot Encoding
Airline = train_data[["Airline"]]
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
Source = train_data[["Source"]]
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
```

```python
Destination = train_data[["Destination"]]
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
data_train = pd.concat([train_data, Airline, Source, Destination], axis = 1)
```

```python
data_train.head()
```

```python
data_train.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)
```

```python
data_train.head()
```

```python
data_train.shape
```

```python
X = data_train.loc[:, ['Total_Stops', 'Journey_day', 'Journey_month', 'Dep_hour',
       'Dep_min', 'Arrival_hour', 'Arrival_min', 'Duration_hours',
       'Duration_mins', 'Airline_Air India', 'Airline_GoAir', 'Airline_IndiGo',
       'Airline_Jet Airways', 'Airline_Jet Airways Business',
       'Airline_Multiple carriers',
       'Airline_Multiple carriers Premium economy', 'Airline_SpiceJet',
       'Airline_Trujet', 'Airline_Vistara', 'Airline_Vistara Premium economy',
       'Source_Chennai', 'Source_Delhi', 'Source_Kolkata', 'Source_Mumbai',
       'Destination_Cochin', 'Destination_Delhi', 'Destination_Hyderabad',
       'Destination_Kolkata', 'Destination_New Delhi']]
X.head()
```

Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python

# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```
